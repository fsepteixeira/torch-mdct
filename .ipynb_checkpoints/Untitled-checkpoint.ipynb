```python
from IPython.display import Audio
import torch, torchaudio
from mdct import MDCT

path = 'samples/original.wav'
wav, sr = torchaudio.load(path)
    
filter_length = int(30e-3*sr)
window_length = int(20e-3*sr)
    
print(filter_length)
print("Wav:", len(wav[0]))

mdct = MDCT(filter_length, window_length, sampling_rate=sr, pad=True, save_pad=True)
rec = mdct.reconstruct(wav)
torchaudio.save('samples/reconstructed.wav', src=rec, sample_rate=sr)

Audio(wav, rate=sr)
```
